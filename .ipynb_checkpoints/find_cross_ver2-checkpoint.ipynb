```python
%matplotlib notebook
%matplotlib inline

import numpy as np
import os
import cv2
import sys
import matplotlib.pyplot as plt
from PIL import Image
import tensorflow as tf
```

```python
ROOT_DIR = os.path.abspath('../find_cross_pattern/')
#sys.path.append(os.path.join(ROOT_DIR,'images/'))
IMG_DIR = os.path.join(ROOT_DIR,'pics\\')
IMG_PATH = os.path.join(IMG_DIR,'test.jpg')

img = cv2.imread(IMG_PATH,1)
[x,y,z]=img.shape
img=cv2.resize(img,(int(y/10),int(x/10)),interpolation=cv2.INTER_CUBIC)
plt.imshow(img)
```

```python
h_low = 6/2
h_high =16/2

s_low = 36 * 255 / 100
s_high = 50 * 255 / 100

v_low = 84 * 255 / 100
v_high = 98 * 255 / 100
```

```python
hsv_img = cv2.cvtColor(img,cv2.COLOR_BGR2HSV)
cv2.imwrite("hsv.jpg",hsv_img)
lower_bound = np.array([h_low,s_low,v_low])
upper_bound = np.array([h_high,s_high,v_high])
hsv_img = cv2.inRange(hsv_img, lower_bound,upper_bound)
plt.imshow(hsv_img)
```

```python
kernel=np.ones((3,3),np.uint8)
opening = cv2.morphologyEx(hsv_img,cv2.MORPH_OPEN,kernel)
plt.imshow(opening)
```

```python
kernel2=np.ones((20,20),np.uint8)
closing = cv2.morphologyEx(opening,cv2.MORPH_CLOSE,kernel2)
plt.imshow(closing)
```

```python
sobelX = cv2.Sobel(closing,cv2.CV_64F,1,0)#x
sobelY = cv2.Sobel(closing,cv2.CV_64F,0,1)#y

sobelX = np.uint8(np.absolute(sobelX))#x
sobelY = np.uint8(np.absolute(sobelY))#y

sobel = cv2.bitwise_or(sobelX,sobelY)
ret,b_sobel = cv2.threshold(sobel,10,20,cv2.THRESH_BINARY)
plt.imshow(b_sobel)
```

```python
canny = cv2.Canny(closing,100,200)
ret,b_canny = cv2.threshold(sobel,5,150,cv2.THRESH_BINARY)
plt.imshow(b_canny)
```

```python
lines = cv2.HoughLines(b_canny,1,np.pi/180,54)
lines1 = lines[:,0,:]#
cnt=0
for rho,theta in lines1[:]: 
    a = np.cos(theta)
    b = np.sin(theta)
    x0 = a*rho
    y0 = b*rho
    x1 = int(x0 + 1000*(-b))
    y1 = int(y0 + 1000*(a))
    x2 = int(x0 - 1000*(-b))
    y2 = int(y0 - 1000*(a)) 
    cv2.line(img,(x1,y1),(x2,y2),(255,0,0),1)
    pair = [x1,y1,x2,y2]
    cnt+=1
    print(pair)
print(cnt)
```
